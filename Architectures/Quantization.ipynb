# Quantization Techniques Notes

## Mixed Precision
Combines different numerical precisions (e.g., FP32, FP16, INT8) in training/inference to balance speed and accuracy. Commonly uses FP16 for computations with FP32 for critical accumulations to prevent overflow/underflow.

## Fine Grained Quantization
Applies quantization at a more detailed level, such as per-channel, per-token, or per-group, rather than globally per layer. This captures finer variations in data distribution, improving accuracy for tasks like LLM inference.

## Increasing Accumulation Precision
Uses higher precision (e.g., FP32) for intermediate accumulations during matrix multiplications or reductions, even if inputs are quantized. This minimizes quantization error buildup in long chains of operations.

## Mantissa Over Exponent
In floating-point quantization, prioritizes preserving mantissa bits (for precision) over exponent bits (for range). This is effective for normalized data where range is limited but precision matters, reducing rounding errors in computations.

## Online Quantization
Dynamically quantizes weights or activations during runtime based on current data statistics, rather than pre-computed scales. Enables adaptation to varying inputs, useful for edge devices or dynamic models, though it adds computational overhead.
